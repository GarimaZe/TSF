###  Using filter Method

```python
df2=df.filter(["Locality","Locality_Verbose","Longitude","Latitude"], axis=1)
df2
```

### Using Transpose Method

```python
df2=df[["Locality","Locality_Verbose","Longitude","Latitude"]].transpose()
df2
```

### using iloc[] method

```python
df3= df.iloc[:, [1,2,3,4]]
df3
```

### Using iloc[] create DataFrame using df.copy() method

```python
df4=df.iloc[:,[1,2,3,4]].copy()
df4
```

### Using loc[] method  
###### https://towardsdatascience.com/how-to-use-loc-in-pandas-49ed348a4117

###### https://sparkbyexamples.com/pandas/pandas-create-new-dataframe-by-selecting-specific-columns/#:~:text=You%20can%20create%20a%20new,added%20to%20the%20original%20ones.

```python
df6=df.loc[:, ["City", "Longitude"]]
df6
```

```python
df7=df.loc[1:3, ["City"]]
df7
```

```python
df8=df.loc[[1,3], ["City"]]
df8
```

```python
df9=df.loc[3:5, "Restaurant_ID":"Address"]
df9
```

```python
df10 = df.loc[df.City=='Makati City', :]
df10
```

```python
df11= df.loc[df.City=='Makati City', "Restaurant_ID":"Address"]
df11
```

```python
df12=df.loc[:, df.columns.drop(["Country_Code"])]
df12
```

```python
df13=pd.DataFrame().assign(Restaurant_ID=df["Restaurant_ID"], Restaurant_Name=df["Restaurant_Name"], Address=df["Address"])
df13
```
